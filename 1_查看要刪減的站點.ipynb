```python
import pandas as pd

# 讀取 CSV 檔案
data = pd.read_csv('data_without_2021.csv')
```

```python
df = pd.DataFrame(data)

# 根據 'siteid' 分組並計算每個 siteid 的記錄數量
siteid_counts = df['siteid'].value_counts().reset_index()
siteid_counts.columns = ['siteid', 'count']

# 設置顯示所有行
pd.set_option('display.max_rows', None)

# 顯示結果
print(siteid_counts)
```

```python
# 縣市和站點的對應關係
locations = {
    "苗栗": [25, 26, 27],
    "南投": [36, 69, 72],
    "彰化": [33, 34, 35, 83],
    "台中": [28, 29, 30, 31, 32],
    "雲林": [37, 38, 41]
}

# 提取特定站點的資料
specific_siteids = [siteid for siteids in locations.values() for siteid in siteids]
filtered_df = df[df['siteid'].isin(specific_siteids)]

# 根據 'siteid' 分組並計算每個 siteid 的記錄數量
siteid_counts = filtered_df['siteid'].value_counts().reset_index()
siteid_counts.columns = ['siteid', 'count']

# 顯示結果
print(siteid_counts)
```

```python
import matplotlib.pyplot as plt

# 定義要繪圖的欄位
columns_to_plot = ['so2', 'o3', 'o3_8hr', 'no2', 'windspeed', 'winddirec', 'co_8hr', 'pm2.5_avg', 'pm10_avg']

# 為每個欄位繪製直方圖
for column in columns_to_plot:
    plt.figure(figsize=(10, 6))
    plt.hist(data[column].dropna(), bins=30, alpha=0.7, color='blue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```
